```python
with open('temp-log.lammps','r') as f:
    logs=f.readlines()
logs=logs[573:]
```

```python
import pandas as pd
df = pd.DataFrame(columns=['Step', 'Temp', 'E_pair','E_mol','TotEng','Press'])
for i,j in enumerate(logs):
    a=j.split()
    if('NNP' in a):
        print(a)
        pass
    else:
        df.loc[i] = a
```

```python
df['Step']= pd.to_numeric(df['Step'], errors='ignore')
df['TotEng']=pd.to_numeric(df['TotEng'], errors='ignore')
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=df['Step'][::500]

y=df['TotEng'][::500]
plt.plot(x[1:], y[1:], alpha=0.5)
plt.title('Scatter plot E v/s Time')
plt.xlabel('Step number - time step = 1fs')
plt.ylabel('Total Energy(eV)')

plt.xticks(np.arange(min(x), max(x)+3000, 3000.0),rotation=80)
plt.show()
```

```python
import pandas as pd
import numpy as np

with open('dataset_1593.xyz','r') as f:
    lines=f.readlines()
frames=[]
temp=[]
for i,line in enumerate(lines):
    if('O' in line or 'H' in line):
        values=line.split()
        temp.append([float(values[4]),float(values[5]),float(values[6])])
    elif('TotEnergy' in line and i > 1):
        #print(i,line)
        frames.append(temp)
        temp=[]
frames=np.array(frames)        
# print(frames)
frames.shape
```

```python
print('frame \t\t force vector \t\t\t force_magnitude')
for j,i in enumerate(frames):
    #print(i)
    total_force=i.sum(axis=0) 
    print(str(j)+"\t"+str(total_force)+"\t"+str(np.linalg.norm(total_force)))
```
